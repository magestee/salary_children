```python
b = 0.2
w = 1

plt.scatter(scaled_x, scaled_y)

axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = b + w * x_vals
plt.plot(x_vals, y_vals, '--', color='red')

plt.show()
```

```python
def compute_cost(x, y, w, b): 

    m = x.shape[0] 
    
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost
```

```python
compute_cost(scaled_x , scaled_y, w, b)
```

```python
w = np.linspace(0, 10, 100)
```

```python
j = np.array([])

for i in w:
    j = np.append(j, compute_cost(scaled_x , scaled_y, i, b))

j
```
